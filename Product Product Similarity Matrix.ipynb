```python
import pandas as pd
import torch
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
products = pd.read_csv("product_info.csv")
```

```python
# create a TfidfVectorizer object
vectorizer = TfidfVectorizer()
```

```python
# define a list of product names
product_names = products.product_name
```

```python
# vectorize the product names
tfidf_matrix = vectorizer.fit_transform(product_names)
```

```python
# compute the pairwise cosine similarity
similarity_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
similarity_df = pd.DataFrame(similarity_matrix,index=product_names,columns=product_names)
```

```python
# Define a list of product ingredients
product_ingredients = products[["product_name","ingredients"]][products.ingredients.notnull()]
ingredients = products.ingredients[products.ingredients.notnull()]
```

```python
# Vectorize the product ingredients
tfidf_matrix_ingredients = vectorizer.fit_transform(ingredients)
```

```python
# Compute the pairwise cosine similarity
similarity_matrix_ingredients = cosine_similarity(tfidf_matrix_ingredients, tfidf_matrix_ingredients)
```

```python
# Create dataframe for similarity matrix
similarity_ingred = pd.DataFrame(similarity_matrix_ingredients,index=product_ingredients.product_name,columns=product_ingredients.product_name)
```

```python
# Define a list of product highlights
product_highlights = products[["product_name","highlights"]][products.highlights.notnull()]
highlights = products.highlights[products.highlights.notnull()]
```

```python
# Create dataframe for similarity matrix
similarity_highlight = pd.DataFrame(similarity_matrix_highlights,index=product_highlights.product_name,columns=product_highlights.product_name)
```

```python
for i in similarity_df.index:
  for j in similarity_df.columns:

    if i==j:
      continue

    retrieve_values = lambda a : a if a.shape==() else a.iloc[0] if len(a.shape)==1  else a.iloc[0,0]

    try:
        product_name_similarity = retrieve_values(similarity_df.loc[i,j])
    except:
        print(i,j,similarity_df.loc[i,j])

    try:
      ingredients_similarity = retrieve_values(similarity_ingred.loc[i,j])
    except:
      ingredients_similarity = -1

    try:
      highlights_similarity = retrieve_values(similarity_highlight.loc[i,j])
    except:
      highlights_similarity = -1

    if ((ingredients_similarity!=-1) & (highlights_similarity!=-1)):
        similarity_df.loc[i,j] =  product_name_similarity*0.4 + ingredients_similarity*0.4 + highlights_similarity*0.2
    elif ingredients_similarity == -1 :
        similarity_df.loc[i,j] = product_name_similarity*0.7 + highlights_similarity*0.3
    elif highlights_similarity == -1:
        similarity_df.loc[i,j] = product_name_similarity*0.6 + ingredients_similarity*0.4
    else:
        similarity_df.loc[i,j] = product_name_similarity
```
